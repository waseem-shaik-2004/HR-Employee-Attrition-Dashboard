```python
# Removing unnecessary columns
df = df.drop(columns=[
    "EmployeeCount",
    "Over18",
    "StandardHours",
    "EmployeeNumber"
])

# Verify the new shape
df.shape
```

```python
df["Attrition"].value_counts()
```

```python
(df["Attrition"] == "Yes").mean() * 100
```

```python
total_employees = len(df)

employees_left = (df["Attrition"] == "Yes").sum()

employees_stayed = (df["Attrition"] == "No").sum()

attrition_rate = (employees_left / total_employees) * 100
```

```python
print(f"Total Employees: {total_employees}")
print(f"Employees Left: {employees_left}")
print(f"Employees Stayed: {employees_stayed}")
print(f"Attrition Rate: {attrition_rate:.2f}%")
```

```python
# Business Observation

# - The organization has a total of **1,470 employees**.
# - **237 employees** have left the company.
# - **1,233 employees** are currently employed.
# - The overall attrition rate is **16.12%**.

# This metric serves as a key HR KPI and provides a baseline for further analysis to identify the factors contributing to employee attrition.
```
